```python
import numpy as np 
import pandas as pd
df = pd.read_csv('employee_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.pie(df['Gender'].value_counts(), labels=df['Gender'].value_counts().index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Experience (Years)'], kde=True)
plt.title('Distribution of Experience (Years)')
plt.xlabel('Experience (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Position', palette='coolwarm',order=df['Position'].value_counts().index)
plt.title('Count of Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data= df.groupby('Position')['Salary'].mean().reset_index(), x='Position', y='Salary', palette='coolwarm')
plt.title('Average Salary by Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(4, 5))
sns.barplot(data=df.groupby('Gender')['Salary'].mean().reset_index(), x='Gender', y='Salary')
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Position', hue='Gender', palette='dark')
plt.title('Gender Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Set the style for plots
sns.set_theme(style="whitegrid")

# Bar Chart: Average salary by position
plt.figure(figsize=(10, 6))
avg_salary_by_position = df.groupby('Position')['Salary'].mean().sort_values()
sns.barplot(x=avg_salary_by_position, y=avg_salary_by_position.index, palette="viridis")
plt.title("Average Salary by Position", fontsize=14)
plt.xlabel("Average Salary")
plt.ylabel("Position")
plt.show()
```

```python

# Scatter Plot: Experience vs. Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Experience (Years)', y='Salary', hue='Gender', palette="coolwarm", alpha=0.7)
plt.title("Experience vs. Salary", fontsize=14)
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.legend(title="Gender")
plt.show()
```

```python

# Heatmap: Correlation matrix
plt.figure(figsize=(6, 4))
correlation_matrix = df[['Experience (Years)', 'Salary']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="Blues", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python

# Box Plot: Salary distribution by gender
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Gender', y='Salary', palette="pastel")
plt.title("Salary Distribution by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.show()
```
